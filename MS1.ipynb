```python
def read_seqres(path_to_file):
    
    
        flag_not_std_aa = False
        not_std_aa = []
        
        with open(path_to_file, "r") as file_pdb:
            lines = file_pdb.readlines()
        
        seqres_list = []
        
        seq_start = 1
        seq_end = 1
        modres_dict = {}
        nstd_dict = {}
        
        flag_comment_dbref = False
        
        
        for line in lines:
            
            if (line[14:18].replace(" ", "")).isdigit():
                if line.startswith("DBREF"):
                    seq_start = int(line[14:18]) - 1
            
                if line.startswith("DBREF"):
                    seq_end = int(line[20:24]) 
                
            else:
                flag_comment_dbref = True
                
            
            if line.startswith("SEQRES"):
                seqres_list.extend(line.split()[4:]) 
                
                
            if line.startswith("MODRES"):
                modres_dict[line[12:15]] = line[24:27]
        
        seqres_dict = {k + seq_start: v for k, v in enumerate(seqres_list, 1)} 
        #seqres_reference = seqres_list.copy() 
        
        for aa in set(seqres_dict.values()):
            if aa not in amino_acids_list:
                flag_not_std_aa = True
                if aa in modres_dict.keys():
                    nstd_dict[aa] = modres_dict[aa]
                else:
                    nstd_dict[aa] = 'info not found in MODRES'
                    
                
        return seqres_dict, seq_start, seq_end, flag_not_std_aa, nstd_dict, flag_comment_dbref
    
    
def read_res(path_to_file, seq_start, seq_end):
    
        with open(path_to_file, "r") as file_pdb:
            lines = file_pdb.readlines()
        
        res_list = {}
        pred_res = 0
        flag_comment = False
        
        for line in lines:
            if line.startswith("ATOM") and int(line[22:26]) != pred_res:

                if (line[22:26].replace(' ', '')).isdigit():
                    
                    gap_size = abs(pred_res - int(line[22:26])) - 1

                    for gap in range(1, gap_size + 1):
                        if pred_res > seq_start:
                            res_list[pred_res + gap] = '-'
                            
                    res_list[int(line[22:26])] = line[17:20]
                    pred_res = int(line[22:26])
                else: 
                    flag_comment = True
                    
            if pred_res != seq_end:
                for i in range (pred_res + 1, seq_end + 1):
                    res_list[i] = '-'
                
        return res_list, flag_comment
    
def seq_compare(path):

    seqres_list, start, end, flag_not_std_aa, not_std_aa, flag_comment_dbref = read_seqres(path)
    res_list, flag_comment = read_res(path, start, end)

    
    return seqres_list, res_list, flag_comment, start + 1, end, flag_not_std_aa, not_std_aa, flag_comment_dbref

def find_gaps(res_dict_original, start, end, cutoff):
    
    k = 0
    pred_res_num = -1
    pred_res_aa = '_'
    gap_start = -1
    gap_dict = {}
    flag_big_gap = False
    
    for res_num, res_aa in res_dict_original.items():
        
        if res_aa == '-':
            k += 1
            if pred_res_aa != '-':
                gap_start = res_num
        else:
            if pred_res_aa == '-' and gap_start >= start and gap_start > 1 and gap_start + k <= end:
                gap_dict[gap_start] = k
                if k >= cutoff:
                    flag_big_gap = True
            k = 0
   
        pred_res_num = res_num
        pred_res_aa = res_aa
    
    return gap_dict, flag_big_gap
```

```python
files_wt = os.listdir(dir_pdb)
files_pdb = list(filter(lambda x: x.endswith('.ent'), files_wt))
```

```python
df = pd.DataFrame(columns = ["Filename", 
                             "Header info (y/n)", 
                             "N of missing residues", 
                             "Missing residues", 
                             "SEQRES (dict)",
                             "Residues (dict)",
                             "SEQRES (len)",
                             "Residues (len)",
                             "Seqres matches pdb res_list",
                             "Starting residue",
                             "Ending residue",
                             "Not standard amino acid",
                             "Not standard amino acid (list)",
                             "Gaps (dict)",
                             "Gap >= 10",
                             "Comment"])





k=0

for filename in tqdm.tqdm_notebook(files_pdb):
    
    #k +=1
    #if k>20:break
    
    
    flag_header = False
    flag_corresp = True
    data_temp = []
    path = dir_pdb + '/' + filename
    dict_missing_res = {}
    comment = 'Ok'
    flag_comment_header = False
    flag_comment_body = False
    flag_not_std_aa = False
    cutoff = 10
    
    with open(path, "r") as handle:
        header_dict = parse_pdb_header(handle)
        handle.close()
        
    if 'has_missing_residues' in header_dict:
        if header_dict['has_missing_residues']:
            
            flag_header = True
            
            dict_missing_res, flag_comment_header = find_missing_residues(path)
            seqres_list, res_list, flag_comment_body, start_seq, end_seq, flag_not_std_aa, not_std_aa, flag_comment_dbref = seq_compare(path)

            N_missing = len(dict_missing_res)
            
            if N_missing != abs(len(seqres_list) - len(res_list) - start_seq) : flag_corresp = False
                        
        else: 

            seqres_list, res_list, flag_comment_body, start_seq, end_seq, flag_not_std_aa, not_std_aa, flag_comment_dbref = seq_compare(path)
            N_missing = abs(len(seqres_list) - len(res_list))
            
    #list_seqres.append(seqres_list)
    #list_res.append(res_list)
    #list_miss.append(dict_missing_res)
    
    if flag_comment_header: 
                comment = 'Error in parsing header'
    if flag_comment_body: 
                comment = 'Error in parsing body'
    
    gap_dict, flag_big_gap = find_gaps(res_list, start_seq, end_seq, cutoff)
    
    data_temp.append({"Filename": filename, 
                      "Header info (y/n)": flag_header, 
                      "N of missing residues": N_missing,
                      "Missing residues": dict_missing_res,
                      "SEQRES (dict)": seqres_list,
                      "Residues (dict)": res_list,
                      "SEQRES (len)": len(seqres_list),
                      "Residues (len)": len(res_list),
                      "Seqres matches pdb res_list": flag_corresp,
                      "Starting residue": start_seq,
                      "Ending residue": end_seq,
                      "Not standard amino acid": flag_not_std_aa,
                      "Not standard amino acid (list)": not_std_aa,
                      "Gaps (dict)": gap_dict,
                      "Gap >= 10": flag_big_gap, 
                      "DBREF parsing": flag_comment_dbref,
                      "Comment": comment}
                    )
    df_temp = pd.DataFrame(data_temp)
    df = df.append(df_temp, ignore_index = True)
    
    
df.to_csv("df.csv ")
```
